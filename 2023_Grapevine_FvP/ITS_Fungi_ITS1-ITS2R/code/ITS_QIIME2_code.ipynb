### Assign taxonomy using a naive bayesian classifier against the UNITE ITS database:

```bash
qiime feature-classifier classify-sklearn \
	--i-classifier unite-ver9-99-classifier-27.10.2022.qza \
	--i-reads Grapevine-ITS-rep-seq.qza \
	--o-classification Grapevine-ITS-UNITE.qza \
    --p-n-jobs 24
```

```bash
qiime metadata tabulate \
    --m-input-file Grapevine-ITS-UNITE.qza \
    --o-visualization Grapevine-ITS-UNITE.qzv
```

### Tidy Table and Filtering

```bash
qiime feature-table filter-samples \
  	--i-table Grapevine-ITS-table.qza \
  	--m-metadata-file metadata/Grapevine-ITS-metadata_tidy.tsv \
  	--o-filtered-table Grapevine-ITS-table_tidy.qza
```

Run 'Decontam.R' code in RStudio for 'contaminant_asvs.tsv' file.

```bash
qiime feature-table filter-features \
  	--i-table Grapevine-ITS-table_tidy.qza \
  	--p-exclude-ids \
  	--m-metadata-file Decontam_output/contaminant_asvs.tsv \
  	--o-filtered-table table1.qza && \

echo 'Done Table 1' && \
# Remove Control Samples

qiime feature-table filter-samples \
  	--i-table table1.qza \
  	--m-metadata-file metadata/Grapevine-ITS-metadata_nocontrols.tsv \
  	--o-filtered-table table2.qza && \

echo 'Done Table 2' && \
# Remove ASVs With <10 Seqs

qiime feature-table filter-features \
  	--i-table table2.qza \
  	--p-min-frequency 10 \
  	--o-filtered-table table3.qza && \

echo 'Done Table 3' && \
# Remove Chloroplast and Mitochondria reads

qiime taxa filter-table \
 	--i-table table3.qza \
 	--i-taxonomy Grapevine-ITS-UNITE.qza \
 	--p-exclude mitochondria,chloroplast \
 	--o-filtered-table Grapevine-ITS-table_filtered.qza && \

echo 'Done Filtering' && \

rm Grapevine-ITS-table_tidy.qza && rm table1.qza && rm table2.qza && rm table3.qza && \

echo 'Directory Cleaned'
```

### Run alpha-rarefaction analysis to see what a good sampling depth will be:

```bash
qiime diversity alpha-rarefaction \
--i-table Grapevine-ITS-table_filtered.qza \
--m-metadata-file metadata/Grapevine-ITS-metadata_nocontrols.tsv \
--o-visualization table-rarefaction.qzv \
--p-min-depth 350 \
--p-max-depth 5000
```

### Fig 6: Feature Volatility Plot

#### q2Longitudinal

First, filter table so only PMA treated samples remain.

```bash
cd QIIME2_output
```

```bash
qiime feature-table filter-samples \
  	--i-table Grapevine-ITS-table_filtered.qza \
  	--m-metadata-file metadata/Grapevine-ITS-metadata_pma.tsv \
  	--o-filtered-table Grapevine-ITS-table_pma.qza
```

Now, collapse the Table file with the Taxonomy file:

```bash
qiime taxa collapse \
	--i-table Grapevine-ITS-table_pma.qza \
	--i-taxonomy Grapevine-ITS-UNITE.qza \
	--p-level 6 \
	--o-collapsed-table Grapevine-ITS-taxtable_pma.qza
```

Note: The directory 'q2Longitudinal' should not exist prior to running the following script:

```bash
qiime longitudinal feature-volatility \
	--i-table Grapevine-ITS-taxtable_pma.qza \
	--m-metadata-file metadata/Grapevine-ITS-metadata_pma.tsv \
	--p-state-column day \
	--p-individual-id-column studyid \
	--p-n-estimators 10 \
	--p-random-state 17 \
	--output-dir q2Longitudinal
```

Output file 'volatility.qzv' 

Plot showing changes to features assigned to the genus 'Erysiphe' which had the highest importance score. Net average change is also shown. Feature list is truncated to show only the top 20 most important features.
